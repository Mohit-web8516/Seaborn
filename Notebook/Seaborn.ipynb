# Seaborn - Introduction

## What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib.

It provides:

- Beautiful default styles
- Statistical visualizations
- Easy integration with Pandas DataFrames
- Less code compared to Matplotlib

Seaborn is widely used for:

- Exploratory Data Analysis (EDA)
- Data Analytics
- Machine Learning Visualization
- Statistical Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Built-in Datasets

One of the biggest advantages of Seaborn is that it provides several built-in datasets for practice.

Some commonly used datasets are:

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins['species'].value_counts()
```

```python
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "island", palette= "Set1")
sns.set_style("whitegrid") #to add background grid 

sns.set_context("poster") # data point size get increase
sns.despine()


```

```python
sns.scatterplot(data = penguins,x = "species" ,y = "body_mass_g",hue = "island", style = "sex", alpha = 0.5)
```

```python
# swarm plot
sns.set_context("talk")
sns.stripplot(data = penguins,x = "species" ,y = "body_mass_g",hue = "island", jitter = True, dodge = True )
```

```python
# histogram
sns.set_context("notebook")
sns.histplot(data = penguins, x ="body_mass_g", hue = "sex",multiple = "stack")
```

```python
#regression plot
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", color = "red")
```

```python
#line plot
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm" , hue = "island", style = "sex")
```

```python
# jointplots

sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

```python

sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "sex", kind = "kde")
```

```python
# bar plot

sns.barplot(data = penguins, x = "species", y = "body_mass_g", hue = 'sex', palette = ["pink","skyblue"], estimator = np.var)
```

```python
# count plot
sns.countplot(data = penguins, x = "species", hue = "island",palette = "Set2")
plt.savefig("images/count_plot.png", dpi = 300, bbox_inches = "tight")
```

```python
# box plot : 

sns.boxplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue = "island",
    palette="Set2"
)
plt.savefig("images/box_plot.png", dpi = 300, bbox_inches = "tight")
plt.title("Bill Length by Species")
plt.show()
```

```python
sns.violinplot(
    data=penguins,
    x = "species",
    y="bill_length_mm",
    hue = "sex"
)
plt.savefig("images/violin_plot.png", dpi = 300, bbox_inches = "tight")
plt.title("Distribution of Bill Length")
# plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
sns.violinplot(data = penguins, x = "species" , y = "body_mass_g", palette = "pastel")
sns.swarmplot(data = penguins, x = "species", y = "body_mass_g", hue = "island", size = 3)
```

```python
# kde plot
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill = True)
plt.savefig("images/kde_plot.png", dpi = 300, bbox_inches = "tight")
```

```python
sns.kdeplot(
    data=penguins,
    x="bill_length_mm"
)
plt.savefig("images/kde_plot.png", dpi = 300, bbox_inches = "tight")
plt.title("Distribution of Bill Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Density")
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

```python
#heatmap 
columns = [
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]

corr_matrix = penguins[columns].corr()

sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = "coolwarm"
)
plt.savefig("images/heatmap.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

```python
#RUG PLOT


sns.rugplot(
    data=penguins,
    y = "body_mass_g", hue = "species",palette = "Set2", height = 0.05
)

plt.savefig("images/rug_plot.png", dpi = 300, bbox_inches = "tight")
plt.show()
```

```python
# pair plot

sns.pairplot (data = penguins, hue = "sex" , palette = "Set2",diag_kind = "kde" )
plt.savefig("images/pair_plot.png", dpi = 300, bbox_inches = "tight")
```

```python
# pair grid = subplot grid for plotting pairwise relationship in a dataset

graph = sns.PairGrid(
    data=penguins,
    hue="sex",
    palette="Set2"
)
plt.savefig("images/PairGrid_plot.png", dpi = 300, bbox_inches = "tight")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) # lower triangle 
graph.map_diag(sns.histplot) # diagnol
graph.add_legend()



```
